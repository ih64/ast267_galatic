```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.integrate as integrate
from astropy.io import ascii
import matplotlib
font = {'size'   : 15}
matplotlib.rc('font', **font)
```

```python
def scheter(L, alpha):
    #set up constants
    h=.7
    Lstar = (10**10) * (h**-2)
    nstar = 0.012*h

    return nstar * np.exp(-L/Lstar)* (L/Lstar)**(alpha)
```

```python
h=.7
Lstar = (10**10) * (h**-2)
nstar = 0.012*h

L = np.linspace(.1,10)*Lstar

models = [scheter(L,-.5)/scheter(Lstar,-.5), scheter(L,-1)/scheter(Lstar,-1),
          scheter(L,-1.25)/scheter(Lstar,-1.25), scheter(L, -2)/scheter(Lstar,-2)]
```

```python
plt.loglog(L, models[0], 'r-')
plt.loglog(L, models[1], 'b-')
plt.loglog(L, models[2], 'g-')
plt.loglog(L, models[3], 'k-')
```

```python
plt.figure(figsize=(10,8))
plt.plot(np.log10(L), np.log(models[0]), label="alpha = -.5")
plt.plot(np.log10(L), np.log(models[1]), label="alpha = -1")
plt.plot(np.log10(L), np.log(models[2]), label="alpha = -1.25")
plt.plot(np.log10(L), np.log(models[3]), label="alpha = -2")
plt.xlabel('Log10(L)')
plt.legend(loc='lower left')
plt.ylabel('Log10($\phi$)')
```

```python
t=ascii.read('galaxiesDM.dat')
z=t['col3']
Mr = t['col6'] - t['col4']
zbin1 = Mr[(z > 0.03) & (z <= 0.05)]
zbin2 = Mr[(z > 0.05) & (z <= 0.07)]
zbin3 = Mr[(z > 0.07) & (z <= 0.09)]
```

```python
plt.figure(figsize=(10,8))
plt.hist(zbin1,bins=np.arange(-25, -16), weights = np.ones(628)/177.0, histtype = 'step',label='0.03 < z <= 0.05')
plt.hist(zbin2,bins=np.arange(-25, -16), weights = np.ones(753)/322.0, histtype = 'step',label='0.05 < z <= 0.07')
plt.hist(zbin3,bins=np.arange(-25, -16), weights = np.ones(983)/467.0, histtype = 'step',label='0.07 < z <= 0.09')
plt.legend(loc='upper left')
plt.xlabel('$M_{r}$')
```
